```python
import numpy as np
from sklearn.model_selection import train_test_split

# Load data
X_dl = np.load("/content/drive/MyDrive/X_eeg.npy")
y = np.load("/content/drive/MyDrive/y_labels.npy")

X_classical = np.load("/content/drive/MyDrive/X_classical.npy")

print("DL shape:", X_dl.shape)
print("Classical shape:", X_classical.shape)
print("Labels shape:", y.shape)
```

```python
unique, counts = np.unique(y, return_counts=True)
print(dict(zip(unique, counts)))
```

```python
Xdl_train, Xdl_test, y_train, y_test = train_test_split(
    X_dl,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

Xc_train, Xc_test, _, _ = train_test_split(
    X_classical,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
# normalizing, which is different for classic aswell as DL

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
Xc_train = scaler.fit_transform(Xc_train)
Xc_test = scaler.transform(Xc_test)
```

```python
mean = Xdl_train.mean()
std = Xdl_train.std()

Xdl_train = (Xdl_train - mean) / std
Xdl_test = (Xdl_test - mean) / std
```

```python
# verify the splitting

print("DL train:", Xdl_train.shape)
print("DL test:", Xdl_test.shape)
print("Classical train:", Xc_train.shape)
print("Classical test:", Xc_test.shape)
```

```python
# freezing the structure.
np.save("/content/drive/MyDrive/Xdl_train.npy", Xdl_train)
np.save("/content/drive/MyDrive/Xdl_test.npy", Xdl_test)
np.save("/content/drive/MyDrive/Xc_train.npy", Xc_train)
np.save("/content/drive/MyDrive/Xc_test.npy", Xc_test)
np.save("/content/drive/MyDrive/y_train.npy", y_train)
np.save("/content/drive/MyDrive/y_test.npy", y_test)
```
